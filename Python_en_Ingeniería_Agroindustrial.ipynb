## Ejemplos de código prácticos

Aquí tienes tres aplicaciones sencillas pero muy útiles en el día a día agroindustrial:

### 1. Balance de Masa Simple

Este código calcula cuánto producto final obtendrás tras un proceso de deshidratación, basado en la humedad inicial y final.

```
# Cálculo de rendimiento en secado
peso_inicial = 500  # kg
humedad_inicial = 0.80  # 80%
humedad_final = 0.15    # 15%

# Sólidos constantes: Peso inicial * (1 - HI) = Peso final * (1 - HF)
peso_final = (peso_inicial * (1 - humedad_inicial)) / (1 - humedad_final)

print(f"De {peso_inicial}kg iniciales, obtendrás {peso_final:.2f}kg de producto seco.")
```

### 2. Clasificación Automática de Calidad

Imagina un sensor que mide el Grado Brix (azúcar) de una fruta. Este script decide si la fruta va para exportación o para jugo.

```
grados_brix = 14.5

if grados_brix >= 13:
    categoria = "Exportación (Premium)"
elif grados_brix >= 10:
    categoria = "Mercado Local"
else:
    categoria = "Procesamiento (Jugos/Mermeladas)"

print(f"La fruta tiene {grados_brix}° Brix. Clasificación: {categoria}")
```

### 3. Monitoreo de Temperatura (Simulado)

Un bucle que revisa una lectura y lanza una alerta si se sale de los rangos de seguridad en un autoclave o cuarto frío.

```
temperatura_actual = 4.5  # Dato que vendría de un sensor
limite_maximo = 5.0

if temperatura_actual > limite_maximo:
    print("¡ALERTA! La temperatura ha excedido el límite de seguridad.")
else:
    print("Temperatura estable. Sistema operando con normalidad.")
```
